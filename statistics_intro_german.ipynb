# Statistische Grundlagen

**Was ist Statistik?**

Statistik „ist die Lehre von Methoden zum Umgang mit quantitativen Informationen“. Sie ist eine Möglichkeit, "eine systematische Verbindung zwischen Erfahrung und Theorie herzustellen“. Unter Statistik versteht man die Zusammenfassung bestimmter Methoden zur Analyse empirischer Daten. Wir unterscheiden zwischen deskriptiver Statistik und Inferenzststatistik:

**Deskriptive Statistik**
 - Beschreibende und zusammenfassend Statistiken 


**Inferenzstatistik**
 - Stichprobendaten werden verwendet um Aussagen über größere       Zusammenhänge zu analysieren.

**Datentypen**

Datentypen bestimmen maßgeblich, welche Methoden und Visualisierungen wir für die Datenanalyse verwenden. Wir unterscheiden zwischen numerischen und Kategorialen Daten. 

***Numerische Daten (Quantitativ)***
 - Kontinuierliche Variablen (Teilbar)
  - Zeit 
  - Geschwindigkeit 

 - Diskrete Variablen (Zählbar)
  - Anzahl der Haustiere 
  - Anzahl der versendeten Packete

***Kategoriale Daten (Qualitativ)***
  - Nominal (Ungeordnet)
   - Binärdaten
   - Qualitative Informationen
   - Aufenthaltsland

  - Ordinal (Geordnet)
   - Fragebogenzustimmung
   - Hierarchie






```python
# Numerische Daten 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from vega_datasets import data
import statistics 

cars = data.cars()
np.mean(cars['Horsepower'])

plt.scatter(x='Horsepower', y='Acceleration', data= cars, alpha=0.5)
plt.show()
```

```python
# Kategoriale Daten

cars['Origin'].value_counts()
sns.countplot(cars['Origin'])
plt.show()
```

# Deskriptive Statistik

Deskriptive Statistik besteht zumeist aus der Aggregation von Werten. Wir fassen große Datenmengen zusammen um eine beschreibende Zahl für eine Datenverteilung zu bekommen. Dabei helfen uns Lage- und Streuungsmaße und die Schiefe einer Verteilung.

**Lagemaße**

- Mittelwerte 
- Median
- Modus
- Häufigkeiten
- Histogramme
- Verteilungen 

```python
np.mean(cars['Miles_per_Gallon'])
```

```python
statistics.median(cars['Miles_per_Gallon'])
```

```python
statistics.mode(cars['Miles_per_Gallon'])
```

**Streuungsmaße**

- Varianz
- Standartabweichung
- Mittlere Standartabweichung
- IQR

```python
sns.boxplot(x='Horsepower', y='Origin', data=cars, hue='Cylinders')
```

**Häufigkeitstabellen & Grafiken**

Eine Datenverteilung 

```python
sns.distplot(cars['Miles_per_Gallon'], bins = 30)
```

# Inferenzstatistik

Die Inferenzstatistik möchte aus Stichprobendaten Informationen generieren, die für eine größere Grundgesamtheit anwendbar sind. Zumeist können wir dafür aber nicht die Stichprobe vergrößern, sondern müssen mit Mitteln der statistischen Inferenz arbeiten. Hierfür gibt es eine vielzahl an Testmethoden.

**Einfache Testmethoden**

- t-Test
- Binominal Test
- Chi-Quadrat Test
- Mann-Whitney U Test
- Wilcoxon Test

**Regressionsanalysen**
- Einfache lineare Regression
- Multiple Regresssion 
- Logistische Regression 

**Korrrelationsanalysen**
 - Korrelationsanalyse nach Pearson 
 - Spearman Rangkorrelation


 **ANOVA**
 - Einfaktorielle ANOVA
 - Zweifaktorielle ANOVA
 - ANOVA mit Messwiederholung



```python
sns.lmplot(x='Displacement', y='Horsepower', data= cars, ci=1)
```
